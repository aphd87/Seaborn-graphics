```python
#Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#View the plot in the notebook
%matplotlib inline
```

```python
#Import auto data
df_auto_data = pd.read_csv('auto_data.csv')
```

```python
#View the top 5 records
df_auto_data.head()
```

```python
#Write a user defined function for origin
# 1-USA, 2-Europe, 3-Asia
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'

#Use apply function
df_auto_data['origin'] = df_auto_data['origin'].apply(origin)
```

```python
#View first 30 data points after applying the user defined function to data set 
df_auto_data.head(30)
```

```python
#Draw the pair. plot using sns for mpg, weight, origin, and with hue origin, set the size to 4
#Note: Hue is variable in data set to map plot aspects to different colors
sns.pairplot(df_auto_data[['mpg','weight','origin']],hue='origin')
```
